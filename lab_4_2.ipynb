```python
from math import *
import ipywidgets as widgets

def findN(i,j):
    return i**2 + i*j + j**2

def findQ(N):
    return sqrt(3* N)

def findSIR(Q, n):
    return (1/6) * pow(Q,n)


def calculateSIR(i,j,n,sir):

    N = findN(i, j)
    print("Currently the cluster size(N) = {}" .format(N))

    Q = findQ(N)
    print("Calculated Frequency Reuse factor(Q) = {}" .format(Q))

    SIR = findSIR(Q, n)
    print("Calculated Signal to interference ratio(SIR) = {}" .format(SIR))

    if(SIR < sir):
        if(i<j):
            i = i+1
        else:
            j = j + 1
            i = 1
        
        print("i= {}" .format(i))
        print("j= {}" .format(j))

        calculateSIR(i, j, n, sir)
    else:
        return (SIR, Q, N)



if __name__ == '__main__':

    i = 1
    j = 2
    sir = 10.0
    # widgets.FloatSlider(value=7.5,min=10.0,max=30.0,step=0.5,description='Required SIR:', orientation='horizontal', readout=True, readout_format='.1f'))
    # widgets.Dropdown(
    # options=[('free space', 2), ('urban area cellular radio', 3.1), ('shadowed urban cellular radio', 4), ('in building LOS', 1.7), ('obstructed in building', 5), ('obstructed in building', 2.5)],
    # description='Choose Environment:',
    # )

    # float_slider = widgets.FloatSlider(value=7.5,min=10.0,max=30.0,step=0.5,description='Required SIR:', orientation='horizontal', readout=True, readout_format='.1f')
    # display(float_slider)
    # sir = float_slider.value
    # print(sir)

    def select_slider(change):
      output.clear_output()
      global sir
      sir = change['new']
      
      print('Required sir: {}'.format(sir))

    slider = widgets.FloatSlider(value=10.0, min=10.0,max=30.0,step=0.5,description='Reqd. SIR:', orientation='horizontal', readout=True, readout_format='.1f')
    output = widgets.Output()
 
    slider.observe(select_slider, names="value")
    display(slider)
    display(output)

    def select(change):
      out.clear_output()
      if(change.new == 'None'):
        with out:          
          display("Select an option")
      elif(change.new == 'free space'):
        with out:
          calculateSIR(i,j,2, sir)
      elif(change.new == 'urban area cellular radio'):
        with out:
          calculateSIR(i,j,3.1, sir)
      elif(change.new == 'shadowed urban cellular radio'):
        with out:
          calculateSIR(i,j, 4, sir)
      elif(change.new == 'in building LOS'):
        with out:
          calculateSIR(i,j, 1.7, sir)
      elif(change.new == 'obstructed in building'):
        with out:
          calculateSIR(i,j,5, sir)
      elif(change.new == 'obstructed in factories'):
        with out:
          calculateSIR(i,j,2.5, sir)
      else:
        with out:
          display("Error: option not found")

    dropdown = widgets.Dropdown(options=['None', 'free space', 'urban area cellular radio', 'shadowed urban cellular radio', 'in building LOS', 'obstructed in building', 'obstructed in factories'])
    out = widgets.Output()
   
    
    dropdown.observe(select, names="value")
    display(dropdown)
    display(out) 

    # drop_down = widgets.Dropdown(
    # options=[('free space', 2), ('urban area cellular radio', 3.1), ('shadowed urban cellular radio', 4), ('in building LOS', 1.7), ('obstructed in building', 5), ('obstructed in building', 2.5)],
    # description='Choose Environment:',
    # )
    # display(drop_down)
    # n = drop_down.value
    # print(n)
    # # print("Considering a seven cell reuse pattern.\n")

    # # calculateSIR(i,j,n, sir)
```
